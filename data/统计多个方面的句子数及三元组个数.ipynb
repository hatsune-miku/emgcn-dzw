```python
import json

with open('/root/autodl-tmp/EMGCN-v2/EMCGCN-v1.1/data/D2/lap14/train.json', 'r') as f:
    data = json.load(f)

aspect_count = {}
triple_count = 0

for item in data:
    triples = item['triples']
    triple_count += len(triples)
    for triple in triples:
        target_tags = triple['target_tags'].split()
        aspect = ""
        for i in range(len(target_tags)):
            if target_tags[i].endswith("\\B"):
                aspect += target_tags[i][:-2]
                break
            elif target_tags[i].endswith("\\I"):
                aspect += target_tags[i][:-2] + " "
            else:
                continue
        if aspect not in aspect_count:
            aspect_count[aspect] = 1
        else:
            aspect_count[aspect] += 1

print("不同方面的句子个数：", len(aspect_count))
print("triples键值对的个数：", triple_count)
print("各方面句子的个数：")
for aspect, count in aspect_count.items():
    print(aspect, count)

print("包含不同方面的句子中的triples的键值对综述：")
for item in data:
    triples = item['triples']
    target_tags = item['target_tags'].split()
    aspect = ""
    for i in range(len(target_tags)):
        if target_tags[i].endswith("\\B"):
            aspect += target_tags[i][:-2]
            break
        elif target_tags[i].endswith("\\I"):
            aspect += target_tags[i][:-2] + " "
        else:
            continue
    if aspect in aspect_count and len(triples) > 0:
        print("句子：", item['sentence'])
        print("triples：")
        for triple in triples:
            print(triple['target_tags'], triple['opinion_tags'], triple['sentiment'])
        print()
```
